# A PARTIR DE AQUÍ EMPIEZA LA MOVIDA DEL RANDOM FOREST

```python
# Library importation
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import pickle
import plotly.express as px
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt
```

```python
# Quitamos la columna date
df_nonan = df.dropna()
df_nodir = df_nonan.drop(columns=['direcc_media_viento'])
df_nodate = df_nodir.drop(columns=['date'])
df_nodate_nostation = df_nodate.drop(columns=['station_id'])

df_nodate_nostation
```

```python
# Create a DataFrame to store the predictions
df_predictions = pd.DataFrame(index=df_nodate_nostation.index)

for column in df_nodate_nostation.columns:
    # Create the input and target datasets
    X = df_nodate_nostation.drop(columns=[column])
    y = df_nodate_nostation[column]

    # Split the data into training and test sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    # Create and train the model
    model = RandomForestRegressor()
    model.fit(X_train, y_train)

    # Make predictions with the model
    y_pred = model.predict(df_nodate_nostation.drop(columns=[column]))

    # Store the predictions in the DataFrame
    df_predictions[column + '_pred'] = y_pred
```

```python
# Crear un diccionario para almacenar los modelos
modelos = {}

for columna in df_nodate_nostation.columns:
    # Crear los conjuntos de datos de entrada y objetivo
    X = df_nodate_nostation.drop(columns=[columna])
    y = df_nodate_nostation[columna]

    # Dividir los datos en conjuntos de entrenamiento y prueba
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    # Crear y entrenar el modelo
    modelo = RandomForestRegressor()
    modelo.fit(X_train, y_train)

    # Almacenar el modelo en el diccionario
    modelos[columna] = modelo

# Save model dictionary
with open('modelos.pkl', 'wb') as f:
    pickle.dump(modelos, f)
```

```python
# Create a dictionary to store the metrics for each column
metrics = {}

for column in df_nodate_nostation.columns:
    # Create the input and target datasets
    X = df_nodate_nostation.drop(columns=[column])
    y = df_nodate_nostation[column]

    # Split the data into training and test sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    # Create and train the model
    model = RandomForestRegressor()
    model.fit(X_train, y_train)

    # Make predictions with the model
    y_pred = model.predict(X_test)

    # Calculate the metrics for this column
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    # Store the metrics in the dictionary
    metrics[column] = {'MSE': mse, 'MAE': mae, 'R^2': r2}

# Now, `metrics` is a dictionary that contains the metrics of each model for each column
```

```python
from docx import Document

# Create a new Word document
doc = Document()

for column, column_metrics in metrics.items():
    # Add the metrics to the Word document
    doc.add_paragraph(f"Metrics for the column {column}:")
    doc.add_paragraph(f"  MSE: {column_metrics['MSE']}")
    doc.add_paragraph(f"  MAE: {column_metrics['MAE']}")
    doc.add_paragraph(f"  R^2: {column_metrics['R^2']}")
    doc.add_paragraph("\n")

# Save the Word document
doc.save('Random_Forest_metrics.docx')
```
